## Binary classification model CNN_LSTM model

```python
import tensorflow as tf
from tensorflow import keras

keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

model = keras.models.Sequential([
                                 keras.layers.Reshape((67,1)), 
                                 keras.layers.Convolution1D(64, kernel_size= 3, strides= 2,padding= 'same', activation= 'relu', input_shape=(67, 1)), 
                                 keras.layers.Convolution1D(64, kernel_size=3, strides=2,padding= 'same', activation= 'relu'), 
                                 keras.layers.MaxPooling1D(), 
                                 keras.layers.Convolution1D(128, kernel_size=3, strides= 2,padding= 'same', activation= 'relu'), 
                                 keras.layers.Convolution1D(128, kernel_size=3, strides= 2,padding= 'same', activation= 'relu'), 
                                 keras.layers.MaxPooling1D(), 
                                 keras.layers.LSTM(70), 
                                 keras.layers.Dropout(0.1), 
                                 keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss= 'binary_crossentropy', metrics= ['accuracy'])
```

```python
early_stopping_cb = keras.callbacks.EarlyStopping(patience = 5, restore_best_weights = True)
```

```python
history = model.fit(train_features_scaled, train_labels, epochs = 50, callbacks = [early_stopping_cb], validation_split= 0.15)
```
